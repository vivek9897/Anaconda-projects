 ## converting timestamp as pandas datetime and extracting date ,time ,month,hour

```python
data['timeStamp'] = pd.to_datetime(data['timeStamp'])
```

```python
data['timeStamp']
```

```python
import datetime as dt
```

```python
data['year'] = data['timeStamp'].dt.year
```

```python
data['month']=  data['timeStamp'].dt.month_name()
```

```python
data['day']=  data['timeStamp'].dt.day_name()
```

```python
data['hour']=  data['timeStamp'].dt.hour
```

```python
data.head()
```

## extracting call type and emergency type values from Title

```python
def calltypesep(x):
    x = x.split(':')
    return x[0]
```

```python

data['call_type']=data['title'].apply(calltypesep)
```

```python
data['call_type']
```

```python
def emergency_typesep(x):
    x = x.split(':')
    return x[1]
```

```python
data['emergency_type'] = data['title'].apply(emergency_typesep)
```

```python
data.head()
```

## visualization

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
font = {'family': 'serif',
        'weight': 'bold',
        'size': 20,
        }
```

```python

plt.bar(x=data.call_type.unique(), height=data.call_type.value_counts())
plt.xlabel('call types', fontdict=font)
plt.xticks(rotation=0)
plt.show()
```

## visualize the percentage share of emergency type for each call 

```python
percentage_share = data.groupby(['call_type', 'emergency_type'])['emergency_type'].count()
```

```python
calls_percentage = percentage_share.groupby(level=0).apply(lambda x: round(100*x/float(x.sum())))
```

```python
calls_percentage
```

```python
sns.set(rc={'figure.figsize':(12,8)})
calls_percentage[:10].plot(kind='bar')
plt.xlabel('emergency_type', fontdict=font)
plt.ylabel('Percentage of Calls', fontdict=font)
plt.xticks(rotation=30)
plt.title('emergency_type/calls_type', fontdict=font)
```

```python
sns.set(rc={'figure.figsize':(15, 7)})
calls_percentage[:10].plot(kind='bar',stacked= True)
```

## calls/months

```python
calls_month = data.groupby(['month','call_type'])['call_type'].count()
```

```python
calls_month
```

```python
calls_month = calls_month.groupby(level=0).apply(lambda x: round(100*x/float(x.sum())))
```

```python
calls_month.head()
```

```python
font = {
    'size': 'x-large',
    'weight': 'bold'
}
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
calls_month = calls_month.reindex(month_order, level=0)
```

```python
calls_month = calls_month.reindex(['EMS', 'Traffic', 'Fire'], level=1)
```

```python
calls_month.head()
```

```python
sns.set(rc={'figure.figsize':(12, 8)})
calls_month.unstack().plot(kind='bar')
plt.xlabel('Name of the Month', fontdict=font)
plt.ylabel('Percentage of Calls', fontdict=font)
plt.xticks(rotation=0)
plt.title('Calls/Month', fontdict=font)
```

## calls/hours

```python
hours_data = data.groupby(['hour', 'call_type'])['call_type'].count()
```

```python
hours_data.head()
```

```python
hours_data_percentage = hours_data.groupby(level=0).apply(lambda x: round(100*x/float(x.sum())))
```

```python
sns.set(rc={'figure.figsize':(18, 8)})
hours_data_percentage.unstack().plot(kind='bar')
plt.xlabel('Hour of the day', fontdict=font)
plt.ylabel('Percentage of Calls', fontdict=font)
plt.xticks(rotation=0)
plt.title('Calls/Hour', fontdict=font)
```
